```python
def run_scored(batsman_name):
    vk = delivery[delivery['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)

print(run_scored('CH Gayle'))
```

```python
#isin()
# find the most destructive death over (khatarnak) batsman in history of ipl 
# strike rate = (number of runs/ number of balls)*100
# min batsman 200 balls is over 16-20
```

```python
delivery.head(3)
```

```python
mask = delivery['over']>15
death_over = delivery[mask]
death_over.head(3)
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
print(all_batsman)

```

```python
x = all_batsman>200
print(x)
```

```python
# now as we studied on numpy the two arrays with same size and one is mask can be applied simply over the other
all_batsman[x]
```

```python
# upper ek series hai and you know in a series if u want to extraact values u write .values and for index u write .index
batsman_list = all_batsman[x].index.tolist()
```

```python
#run made by this batsmans 
#balls played by this batsman
mask = delivery['batsman'].isin(batsman_list)
# deliver[mask]
runs = delivery[mask].groupby('batsman')['batsman_runs'].sum()

balls = delivery[mask].groupby('batsman')['batsman_runs'].count()

ans = (runs/balls)*100
ans

```
