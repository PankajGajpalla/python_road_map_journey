```python
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d') #d stands for integers
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis') 
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=True, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score=0, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, metric='correlation', annot=True)
```

```python
sns.clustermap(x, z_score=0, metric='correlation', row_cluster=False, annot=True)
```
