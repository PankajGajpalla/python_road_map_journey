```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', estimator=np.median, kind='bar', data=tips) #just provide the name of the funcion np.median not to call it like np.median()
```

countplot are special type of barcount its similar to histogram

```python
sns.catplot(x='sex', kind='count', data=tips)
```

```python
sns.catplot(x='sex', hue='smoker', kind='count', data=tips)
```
