```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('salary_age_data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python
#the third value provides where that point is croseed by filling bellow and above it lets see:)

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,where=(py_salaries > overall_median), alpha=0.25)

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,where=(py_salaries > overall_median),interpolate=True, alpha=0.25)

plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
#check the difference the interpolete covers all the area small area that is left help to fill completely
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,where=(py_salaries > overall_median), color="green",interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salaries, overall_median,where=(py_salaries <= overall_median),color= 'red',interpolate=True, alpha=0.25)


plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All_Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,where=(py_salaries > dev_salaries), color="green",interpolate=True, alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries,where=(py_salaries <= dev_salaries),color= 'red',interpolate=True, alpha=0.25, label='Below Avg')


plt.legend()
plt.title('Median Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.tight_layout()
plt.show()
```
