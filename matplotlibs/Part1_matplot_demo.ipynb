```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
dev_x = np.arange(25,36)

dev_y = np.random.randint(38000, 75000,11)

dev_y = np.sort(dev_y)

print(dev_x)
print(dev_y)
```

```python
plt.plot(dev_x, dev_y)
plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
py_dev_y = np.sort(np.random.randint(40000, 80000,11))
plt.plot(dev_x, py_dev_y)
plt.show()
```

```python
plt.plot(dev_x, dev_y)
plt.plot(dev_x, py_dev_y)

plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
plt.plot(dev_x, dev_y)
plt.plot(dev_x, py_dev_y)
plt.legend(['All_devs', 'python']) #as we are passing it on our own it may be an error by putting wrong so see next cell

plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
plt.plot(dev_x, dev_y, label='All_devs')
plt.plot(dev_x, py_dev_y, label='Python')
plt.legend() #put the labels in the ploting it self  and then call the legend:)

plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
plt.plot(dev_x, dev_y, 'k--' ,label='All_devs') # k means black and -- means this type of patter
plt.plot(dev_x, py_dev_y, 'g' ,label='Python') # this is a direct way to do it g means green
plt.legend() 


plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
# the above thing can also be done more readable like 
plt.plot(dev_x, dev_y, marker='.',linestyle='--', color='#444444' ,label='All_devs') #can also use #color values rgb
plt.plot(dev_x, py_dev_y,color='b', marker='o'  ,label='Python') 
plt.legend() 


plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```

```python
js_dev_y = np.sort(np.random.randint(30000, 80000,11))
js_dev_y
```

```python
plt.plot(dev_x, dev_y, marker='.',linestyle='--', color='#444444', linewidth=3, label='All_devs') #can also use #color values rgb
plt.plot(dev_x, py_dev_y,color='b', marker='o', linewidth=3, label='Python') 

plt.plot(dev_x, js_dev_y, color='#adad3b',linewidth=3, label='Javascript') #linewidth to make it bold
plt.legend() 
plt.show()
```

```python
plt.plot(dev_x, dev_y, marker='.',linestyle='--', color='#444444', linewidth=3, label='All_devs') #can also use #color values rgb
plt.plot(dev_x, py_dev_y,color='b', marker='o', linewidth=3, label='Python') 
plt.plot(dev_x, js_dev_y, color='#adad3b',linewidth=3, label='Javascript')

plt.tight_layout() #to correct the padding of the chart 
plt.grid(True) #tonshow grid


plt.legend() 
plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median salary (usd) by Age')
plt.show()
```
