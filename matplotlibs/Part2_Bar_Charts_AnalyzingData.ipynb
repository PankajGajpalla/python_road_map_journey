```python
import csv
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    row = next(csv_reader)

print(row)
print(row['LanguagesWorkedWith'])
print(row['LanguagesWorkedWith'].split(';'))
```

```python
# Counter : python have a special function by useing whihch we can keep tracck of how many times things appear 
from collections import Counter

c = Counter(['Python', 'JavaScript'])
print(c)
c.update(['Python', 'C++'])
print(c)
c.update(['C++', 'Python'])
print(c)
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

print(language_counter)
print()
print(language_counter.most_common(10))
```

```python
languages = []
popularity = []

#lanaguage_counter is  a dictionary and language_counter.most_common(10) is  list of tuple and we r u using this most_comm...

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)

plt.bar(languages, popularity)

plt.title('Most popular languages')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.show()
```
