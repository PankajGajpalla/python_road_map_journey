```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

slices = [60, 40]
labels = ["Sixty", "Forty"]


plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}) #edgecolor is for the edges to :)

plt.title('pieee')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [60, 40, 30, 20]
labels = ["Sixty", "Forty", "Extra1", "Extra2"]
colors = ["#008fd5", "green", "yellow", "red"] #u can use #colors to make it look more professinoal 


plt.pie(slices, labels=labels,colors=colors, wedgeprops={'edgecolor': 'black'}) 

plt.title('pieee')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] #how much away from the radius 0.1 is 10% away see
# explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'}) 

plt.title('pieee')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] 
# explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})  #shadow = true only for asthetic look

plt.title('pieee')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] 
# explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={'edgecolor': 'black'})  #startangle rotates the pi by some angle

plt.title('pieee')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0] 
# explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})  #shows value of percent in the pichart

# plt.legend() 
plt.title('pieee')
plt.tight_layout()
plt.show()
```
