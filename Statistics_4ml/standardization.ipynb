Standardization

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# generates 1000 normal random interges with specified mean and std 

draw = norm.ppf(np.random.random(100000), loc=1000, scale=200).astype(int)
draw
```

```python
sns.distplot(draw)
```

```python
Z = []
for i in draw:
    Z.append((i-1000)/200) #subtract with mean and divide with standard deviation
```

```python
a = np.array(Z)
```

```python
print(a.mean())
print(np.round(a.mean()))
```

```python
# so as you can see above the mean is zero
# whenerver we standardize or convert to standard normal varient then mean become zero 
```

```python
#similarly the std becomes 1

print(a.std())
print(np.round(a.std()))
```

```python
sns.distplot(a)
```

Central Limit Theorm Proof

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
delivery.head(3)
```

```python
runs = delivery.groupby('batsman').sum()['batsman_runs']
runs
```

```python
runs = runs.values
print(runs)
```

```python
sns.distplot(runs)
```

we can see this is not a normal distribution

```python
# now we are going to do sampling with this value 1000 times

np.random.choice(runs, size=10)   #it chose 10 values randomly from the runs 
```

```python
m = 1000
n = 50

new=[]

for i in range(m):
    x = np.random.choice(runs, size=n)
    new.append(x)
```

```python
len(new)
```

```python
new = np.array(new)
new
```

```python
sns.distplot(new)
```
