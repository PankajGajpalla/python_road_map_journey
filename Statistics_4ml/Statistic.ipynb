Cumulative frequency(CF) :
keep adding up all the prev frequency


```python
a = data['pclass'].value_counts()
a

```

```python
a.sum()
```

```python
(a/a.sum())*100
```

```python
# since its a series we can fetch value by useing the .value
b = ((a/a.sum())*100).values
b
```

```python
c = []
cf = 0
for i in b:
    cf += i
    c.append(cf)
c
```

```python
plt.plot(c, marker='D')
```
