## Text Cleaninng

1. Samples 10,000 rows
2. Remove html tags
3. Remove special charecters
4. Coverting every thing in lower case
5. Removing Stop words
6. Stemming

```python
df = df.sample(10000)
```

```python
df.shape
```

```python
df.info()
```

```python
 # replacing setiment positive negative with 0 and 1

df['sentiment'].replace({'positive':1,'negative':0}, inplace=True)
```

```python
df.head()
```

```python
# to remove html tag  we will use regular expression

import re
clean = re.compile('<.*?>')
re.sub(clean, '', df.iloc[2].review)
```

```python
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review'] = df['review'].apply(clean_html)
```

```python
# function to convert to lowercase
def convert_lower(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(convert_lower)
```

```python
# function to remove special charecter

def remove_special(text):
    x = ''

    for i in text:
        if i.isalnum():
            x = x+i
        else:
            x = x+ ' '

    return x
```

```python
remove_special("although @ %% $fdg %  russell was and is probably the closest person to look like elvis in show business  so")
```

```python
df['review'] = df['review'].apply(remove_special)
```

```python
# Remove the stop words
import nltk #natural language tool kit
```

```python
from nltk .corpus import stopwords
```

```python
stopwords.words('english')
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    return [word for word in text.split() if word not in stop_words]
    # x = []
    # for i in text.split():
    #     if i not in stopwords.words('english'):
    #         x.append(i)
    # y = x[:]
    # x.clear()
    # return y
            
```

```python
text = "This is an example sentence showing off the stopwords filtration."
print(remove_stopwords(text))
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
# perform stemming ( stemming -> play playing played = play)
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
# y = []
def remove_stemming(text):
    return [ps.stem(word) for word in text]
    # for i in text:
    #     y.append(ps.stem(i))
    # Z = y[:]
    # y.clear
    # return Z
```

```python
words = ["running", "flies", "easily", "fairly", "playing", "played", "play"]
print(remove_stemming(words))
```

```python
df['review'] = df['review'].apply(remove_stemming)
```

```python
def join_back(list_input):
    return " ".join(list_input)
```

```python
df['review'] = df['review'].apply(join_back)
```
