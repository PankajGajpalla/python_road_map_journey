```python
soup = BeautifulSoup(webpage, 'lxml')

for j, i in enumerate(soup.find_all('tr')[1:26], start=1): #imp note: its in order i and j first is indexing and i is looping over element of soup
    if(j%16==0):
        continue
    print(i.find_all('td', class_='text')[1].text.strip())
```

```python
name = []
P_E = []
market_cap = []
div_yld = []

for j, i in enumerate(soup.find_all('tr')[1:], start=1): #imp note: its in order i and j first is indexing and i is looping over element of soup
    if(j%16==0):
        continue
    name.append(i.find_all('td', class_='text')[1].text.strip())
    P_E.append(i.find_all('td')[3])
    market_cap.append(i.find_all('td')[4])
    div_yld.append(i.find_all('td')[5])

d = {'name':name, 'P_E':P_E, 'market_cap':market_cap, 'div_yld':div_yld}

df=pd.DataFrame(d)
df
```

```python
final = []


for index in range(1, 100):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }

    webpage = requests.get('https://www.screener.in/screens/357649/all-listed-companies/?page={}'.format(index), headers=headers).text

    soup = BeautifulSoup(webpage, 'lxml')

    name = []
    P_E = []
    market_cap = []
    div_yld = []

    for j, i in enumerate(soup.find_all('tr')[1:], start=1): #imp note: its in order i and j first is indexing and i is looping over element of soup
        if(j%16==0):
            continue
        name.append(i.find_all('td', class_='text')[1].text.strip())
        P_E.append(i.find_all('td')[3].text.strip())
        market_cap.append(i.find_all('td')[4].text.strip())
        div_yld.append(i.find_all('td')[5].text.strip())

    d = {'name':name, 'P_E':P_E, 'market_cap':market_cap, 'div_yld':div_yld}

    df=pd.DataFrame(d)
    
    final.append(df)

# Concatenate all DataFrames once (better than appending in a loop)
final_df = pd.concat(final, ignore_index=True)

print(final_df)
final_df.to_csv('stocks_n')
    
```
